## Задача 3

Дан массив. Найти наиболее часто встречающееся значение в массиве и вывести его.

```python
a = np.random.randint(0,10,30)
print(a)
```

```python
# Ваше решение
dic = {}
max_value = 0

for value in a:
    try:
        if dic[value] >= 0:
            dic[value] += 1
    except KeyError:
        dic[value] =1

        
max_value = max(dic, key=dic.get)
 
print("Самое частое значение = ",max_value)
```

## Задача 6

Создайте матрицу 8х8 и заполнить её в шахматном порядке нулями и единицами.

```python
# Ваше решение
matrix = np.zeros((8,8))
for i in range(8):
    for j in range(8):
        if (i%2 ==0) and (j%2 != 0):
            matrix[i,j] = 1
            matrix[j,i] = 1
print(matrix)
```
